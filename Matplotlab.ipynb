```python
company=['google','Amzn','Fb']
rev=[60,90,136]
pro=[40,32,12]
xpos=np.arange(len(company))
plt.bar(xpos-0.2,rev,width=0.4,label='rev')
plt.bar(xpos+0.2,pro,width=0.4,label='pro')

plt.xticks(xpos,company)
plt.ylabel("rev(Bln)")
plt.title("Technology stocks")
plt.legend()
plt.show()
```
